### **Read from CSV file**


```python

# Path to your CSV folder — adjust this!
path = r"Raw Data"

# Get all chunk_*.csv files
# files = glob.glob(f"{path}/november_chunk_*.csv")
files = glob.glob(f"{path}/october_chunk_*.csv")

# Read and combine all files
df_list = []

for file in files:
    try:
        temp = pd.read_csv(file)
        
        # Make sure the date column is parsed as datetime
        if 'date' in temp.columns:
            temp['date'] = pd.to_datetime(temp['date'], errors='coerce')
        else:
            print(f"Warning: 'date' column missing in {file}")
        
        df_list.append(temp)
    except Exception as e:
        print(f"Error reading {file}: {e}")

# Merge everything into one dataframe
full_df = pd.concat(df_list, ignore_index=True)

# Filter by date range
mask = (full_df['date'] >= "2024-10-19") & (full_df['date'] <= "2024-11-11")
filtered_df = full_df.loc[mask]

filtered_df
```
